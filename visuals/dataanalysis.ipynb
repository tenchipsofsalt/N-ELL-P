```python
df = pd.read_csv('/kaggle/input/feedback-prize-english-language-learning/train.csv')
```

```python
scoring_cols = ["cohesion", "syntax", "vocabulary", "phraseology", "grammar", "conventions"]
```

```python
def display_essay_info(row):
    print("Essay:")
    print(row["full_text"])
    print("\n")
    print("Scores:")
    print(row[scoring_cols])
```

```python
rand_row = randrange(df.shape[0])
display_essay_info(df.iloc[rand_row])
```

```python
def int_to_grid_coord(k, n):
    return (k // n) + 1, (k % n) + 1
```

```python
fig = subplots.make_subplots(
    rows=2, cols=3,
    subplot_titles=scoring_cols 
)

for k, scoring_cols  in enumerate(scoring_cols ):
    i, j = int_to_grid_coord(k, 3)
    
    score_mean = df[scoring_cols ].mean()
    score_std = df[scoring_cols ].std()
    
    fig.add_trace(
        go.Histogram(
            x=df[scoring_cols ], nbinsx=9, 
            name=scoring_cols 
        ),
        row=i, col=j
    )
    
    fig.add_shape(
        type='line', 
        x0=score_mean, x1=score_mean, 
        y0=0, y1=1000,
        line=dict(dash='dash', color='black'),
        row=i, col=j
    )
    
    fig.add_shape(
        type='line', 
        x0=score_mean - score_std, x1=score_mean - score_std, 
        y0=0, y1=1000,
        line=dict(dash='dash', color='dark gray'),
        row=i, col=j
    )
    
    fig.add_shape(
        type='line', 
        x0=score_mean + score_std, x1=score_mean + score_std, 
        y0=0, y1=1000,
        line=dict(dash='dash', color='dark gray'),
        row=i, col=j
    )
    
    fig.add_annotation(
        x=score_mean,
        y=-100,
        text="Mean = {:,.2f}".format(score_mean),
        showarrow=False,
        row=i, col=j
    )
    
    fig.add_annotation(
        x=score_mean,
        y=-200,
        text="Std = {:,.2f}".format(score_std),
        showarrow=False,
        row=i, col=j
    )

fig.update_layout(
    title='Score Distributions',
    height=800
)
    
fig.show()
```
